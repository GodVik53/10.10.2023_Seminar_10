# Задача №67.
# 1. Создать новый столбец в таблице с пингвинами, который будет отвечать запоказатель длины клюва пингвина.
#high - длинный(от 42)
#middle - средний(от 35 до 42)
#low - маленький(до 35)

```python
penguins = sns.load_dataset("penguins")
penguins.sample(10)
```

```python
penguins.loc[penguins["bill_length_mm"] >=42, "height_group"] = "high"
penguins.loc[(penguins["bill_length_mm"] >=35) & (penguins["bill_length_mm"] <42), "height_group"] = "middle"
penguins.loc[penguins["bill_length_mm"] <35, "height_group"] = "low"
```

```python
penguins.sample(20)
```

# Задача №69. Изобразить гистограмму по flipper_length_mm с оттенком height_group. Сделать анализ


```python
plt.figure(figsize=(14,6))
sns.histplot(penguins, x="flipper_length_mm", hue="height_group", bins=30) # bins - количество столбцов
plt.title("Размер крыла и клюва")
plt.xlabel("Размер крыла")
plt.ylabel("Колличество")
plt.xticks(range(170,240,2)) # показывает числовые значения с шагом 2 по оси Х
plt.yticks(range(0,27))# показывает числовые значенияпо оси y
plt.grid(); # сетка
```
